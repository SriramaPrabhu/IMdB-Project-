```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options


chrome_options = Options()
chrome_options.add_argument('--headless')
```

```python
database=pd.read_csv('IMdB data project.csv')
```

```python
database.head(4)
```

```python
urlbase=database['MovieLink'][2500:5000]
urlbase
```

```python

def getdata(i):
    d={}
    driver = webdriver.Chrome()
    driver.maximize_window()
    driver.get(i)
    
    try:
        Name=driver.find_element(By.XPATH,"//*[@class='sc-afe43def-1 fDTGTb']")
        d['Name']=Name.text
    except Exception as msg:
        pass
    
    try:
        Certification_Rating=driver.find_element(By.XPATH,"/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[2]/a")
        d['Certification_Rating']=Certification_Rating.text
    except Exception as msg:
        pass
    
    
    try: 
        Run_Time=driver.find_element(By.XPATH,"/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[3]")
        d['RT']=Run_Time.text
    except Exception as msg:
        pass
    
    try:
        Genre=driver.find_element(By.XPATH,"//*[@data-testid='genres']")
        d['Genre']=Genre.text
    except Exception as msg:
        pass
    
    try:
        Plot=driver.find_element(By.XPATH,"//*[@data-testid='plot']")
        d['Plot']=Plot.text
    except Exception as msg:
        pass
    
    try:
        User_Reviews=driver.find_element(By.XPATH,"/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[2]/ul/li[1]/a/span/span[1]")
        d['User_Reviews']=User_Reviews.text
    except Exception as msg:
        pass
    
    try:
        Critic_Reviews=driver.find_element(By.XPATH,"/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[2]/ul/li[2]/a/span/span[1]")
        d['Critic_Reviews']=Critic_Reviews.text
    except Exception as msg:
        pass
    
    try:
        MetaScore=driver.find_element(By.XPATH,"/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[2]/ul/li[3]/a/span/span[1]/span")
        d['MetaScore']=MetaScore.text
    except Exception as msg:
        pass
    
    try:
        Oscar=driver.find_element(By.XPATH,"//*[@data-testid='award_information']")
        d['Oscar']=Oscar.text
    except Exception as msg:
        pass
    
    

    directorlist=[]
    writerlist=[]
    starlist=[]
    try:
        Director=driver.find_element(By.XPATH,'//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[2]/div/ul/li[1]/div/ul')
        director_elements=Director.find_elements(By.TAG_NAME, "li")
        n=0
        for dir_element in director_elements:
            directorlist.append(dir_element.text)
            n=n+1
        if n==1:
            d['Director 1']=directorlist[0]
        elif n==2:
            d['Director 1']=directorlist[0]
            d['Director 2']=directorlist[1]
        elif n==3:
            d['Director 1']=directorlist[0]
            d['Director 2']=directorlist[1]
            d['Director 3']=directorlist[2]
        else:
            d['Director 1']=directorlist[0]
            d['Director 2']=directorlist[1]
            d['Director 3']=directorlist[2]
    except Exception as msg:
        pass
    

    try:
        Writer=driver.find_element(By.XPATH,'//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[2]/div/ul/li[2]/div/ul')
        writer_elements=Writer.find_elements(By.TAG_NAME, "li")
        w=0
        for writer_element in writer_elements:
            writerlist.append(writer_element.text)
            w=w+1
        if w==1:
            d['Writer 1']=writerlist[0]
        elif w==2:
            d['Writer 1']=writerlist[0]
            d['Writer 2']=writerlist[1]
        elif w==3:
            d['Writer 1']=writerlist[0]
            d['Writer 2']=writerlist[1]
            d['Writer 3']=writerlist[2]
        else:
            d['Writer 1']=writerlist[0]
            d['Writer 2']=writerlist[1]
            d['Writer 3']=writerlist[2]
    except Exception as msg:
        pass
        
    try:
        stars=driver.find_elements(By.XPATH,"//*[@class='ipc-inline-list ipc-inline-list--show-dividers ipc-inline-list--inline ipc-metadata-list-item__list-content baseAlt']")
        star_elements = stars[2].find_elements(By.TAG_NAME, 'li')
        s=0
        for star_element in star_elements:
            starlist.append(star_element.text)
            s=s+1
        if s==1:
            d['Star 1']=starlist[0]
        elif s==2:
            d['Star 1']=starlist[0]
            d['Star 2']=starlist[1]
        elif s==3:
            d['Star 1']=starlist[0]
            d['Star 2']=starlist[1]
            d['Star 3']=starlist[2]
        else:
            d['Star 1']=starlist[0]
            d['Star 2']=starlist[1]
            d['Star 3']=starlist[2]
    except Exception as msg:
        pass
    
    try:
        Release_Date=driver.find_element(By.XPATH,"//*[@data-testid='title-details-releasedate']")
        d['Release_Date']=Release_Date.text
    except Exception as msg:
        pass
    
    try:
        Country_Origin=driver.find_element(By.XPATH,"//*[@data-testid='title-details-origin']")
        d['Country_Origin']=Country_Origin.text
    except Exception as msg:
        pass
    
    try:
        Language=driver.find_element(By.XPATH,"//*[@data-testid='title-details-languages']")
        d['Language']=Language.text
    except Exception as msg:
        pass
    
    try:
        Shooting_Locations=driver.find_element(By.XPATH,"//*[@data-testid='title-details-filminglocations']")
        d['Shooting_Locations']=Shooting_Locations.text
    except Exception as msg:
        pass

    try:
        Production_Company=driver.find_element(By.XPATH,"//*[@data-testid='title-details-companies']")
        d['Production_Company']=Production_Company.text
    except Exception as msg:
        pass
    
    try:
        Budget=driver.find_element(By.XPATH,"//*[@data-testid='title-boxoffice-budget']")
        d['Budget']=Budget.text
    except Exception as msg:
        pass
    
    try:
        Boxoffice_Collection=driver.find_element(By.XPATH,"//*[@data-testid='title-boxoffice-cumulativeworldwidegross']")
        d['Boxoffice_Collection']=Boxoffice_Collection.text
    except Exception as msg:
        pass

    try:
        Soundmix=driver.find_element(By.XPATH,"//*[@data-testid='title-techspec_soundmix']")
        d['Soundmix']=Soundmix.text
    except Exception as msg:
        pass
    
    try:
        Aspect_Ratio=driver.find_element(By.XPATH,'//*[@id="__next"]/main/div/section[1]/div/section/div/div[1]/section[14]/div[2]/ul/li[3]/div/ul/li/span')
        d['Aspect_Ratio']=Aspect_Ratio.text
    except Exception as msg:
        pass
    driver.quit()
    return d 

cinemadata=[]
url_num=0
start_time = time.time()
for i in urlbase:
    print(f'started url number {url_num}')
    url_num+=1
    url_content = getdata(i)
#     print(url_content)
    cinemadata.append(url_content)    
    print("URL Completed")
    time.sleep(5)
end_time = time.time()
Time_taken = end_time - start_time
print("time taken to scrape:", Time_taken)
```

```python
Datade=pd.DataFrame(cinemadata,columns=['Name','Certification_Rating','RT','Genre','Plot','User_Reviews','Critic_Reviews','MetaScore','Oscar','Director 1','Director 2','Director 3','Writer 1','Writer 2','Writer 3','Star 1','Star 2','Star 3','Release_Date','Country_Origin','Language','Shooting_Locations','Production_Company','Budget','Boxoffice_Collection','Soundmix','Aspect_Ratio'])
```

```python
Totaldata=pd.concat([Datade])
Totaldata.to_excel('IMDB Project total data.xlsx')
```
